Q2: Demonstrate Lists and Dictionary comprehensions

Write a PYTHON script with List comprehension for the following
• Is the given list divisible by 3 or NOT?
• Square of even numbers in a list
• Sum of digits of all EVEN numbers in a list
• Remove duplicate numbers in a list

```python
#1 List 

#Is the given list divisible by 3 or NOT?
def divisible(pax_capacity):
    return [num for num in pax_capacity if num % 3 == 0]

#Square of even numbers in a list
def square(pax_capacity):
    return [num**2 for num in pax_capacity if num % 2 == 0]

#Sum of digits of all EVEN numbers in a list
def sum_even(pax_capacity):
    return [sum(int(digit) for digit in str(num)) for num in pax_capacity if num % 2 == 0]

#Remove duplicate numbers in a list
def duplicates(pax_capacity):
    return list(set(pax_capacity)) # as set in python cannot have duplicate values

def menu():
    print("\n-------------Welcome to Q2 LAB Python-------------------")
    print("\n---------------------New Menu-------------------")
    print("\n\[1] Is the given list divisible by 3 or NOT?")
    print("[2] Square of even numbers in a list")
    print("[3] Sum of digits of all EVEN numbers in a list")
    print('[4] Remove duplicate numbers in a list')
    print("[0] Exit the program.")
    


menu()

option = int(input("Enter option: "))
 
pax_capacity = [320, 50,56,182,480,85,12,90,12,45, 320, 50]

while option != 0:
    if option == 1:
        print("\nIs the given list divisible by 3 or NOT?")
        print(divisible(pax_capacity))
 
    elif option == 2:
        print("\nSquare of even numbers in a list")
        print(square(pax_capacity))
 
    elif option == 3:
        print("\nSum of digits of all EVEN numbers in a list")
        print(sum_even(pax_capacity))

    elif option == 4:
        print("\nRemove duplicate numbers in a list")
        print(duplicates(pax_capacity))

    else:
        print("Invalid option.")
     
    menu()                  
    option = int(input("Enter option : "))
 
print("Thank you, You can Exit")
```
